```python
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris,load_boston
from IPython.display import display
import pandas as pd
import numpy as np
```

```python
boston = load_boston()
boston_data = boston.data
boston_target = boston.target
#for x in range(boston_data.shape[1]):
#    print(boston_data[x].dtype)

b_frame = pd.DataFrame(boston_data,columns=boston.feature_names)
#display(b_frame.head())

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
y2 = np.cos(x **2 )
```

```python
### Docs
#http://matplotlib.org/api/axes_api.html#axis-labels-title-and-legend

#most basic plot
x_vals=range(10)
y_vals=np.array(range(10))+4

plt.plot(x_vals,y_vals)
plt.show()
```

```python

#subplots
#A figure object and and axes array
# 2 rows and 3 column
plt.clf()
figure, axes = plt.subplots(2,3)
count =0
for i in range(2):
    for j in range(3):
        axes[i,j].scatter(boston_data[:,count],boston_target)
        axes[i,j].set_title("{} vs rent".format(boston.feature_names[count]))
        count+=1
plt.tight_layout()
plt.show()





```

```python
#change the size of our plots
#width height in inches, but uses dpi value to display
plt.clf()
plt.figsize=((15,20))
plt.scatter(b_frame['RM'],boston_target)
plt.show()
```

```python
#legends and title 
#label value appears automatically as a legend
#when .legend() is called
plt.clf()
plt.figsize=((15,20))
plt.plot(x,y,label="Y")
plt.plot(x,y2,label="Y2")
plt.title("some aribitrary data")
plt.legend()
plt.show()


#saving to a file
plt.savefig("arbitrary.png",format="png")
```

```python
#plain bar
bar_data=np.array([5,2,10,25,16])
plt.bar(range(len(bar_data)),bar_data)
plt.show()
```

```python
#rotated bar
bar_colors =["red","blue","green","yellow","black"]
plt.barh(range(len(bar_data)),bar_data,color=bar_colors)
plt.show()
```
